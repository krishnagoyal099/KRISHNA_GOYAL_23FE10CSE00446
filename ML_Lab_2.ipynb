```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# 1. Choose a valid file from the dataset:
# Options: "Red.csv", "White.csv", "Rose.csv", "Sparkling.csv", "Varieties.csv"
file_path = "Red.csv"

# 2. Load the specific file
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "budnyak/wine-rating-and-price",
  file_path,
)

print(f"Successfully loaded {file_path}")
print("First 5 records:", df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Remove duplicates and missing values
df = df.drop_duplicates().dropna()

# One-hot encode categorical variables
df = pd.get_dummies(df, drop_first=True, dtype=float)

# Scale features
scaler = StandardScaler()
df[df.columns] = scaler.fit_transform(df)

print(df.head())
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "budnyak/wine-rating-and-price",
    "Red.csv"
)

# 2. Preprocess (Select numeric to avoid errors/memory issues)
df = df.select_dtypes(include=['number']).dropna()

# 3. Plot
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap='coolwarm', center=0, annot=True)
plt.show()
```
